```python
df['Volume'].plot()
plt.tight_layout()
plt.grid()
```

```python
average_open = df['Open'].mean()
average_close= df['Close'].mean()
average_high = df['High'].mean()
average_low = df['Low'].mean()
```

```python
averages = {
    'Average Open': average_open,
    'Average Close': average_close,
    'Average High': average_high,
    'Average Low': average_low
}

# Create bar plot
plt.figure(figsize=(10, 6))
plt.bar(averages.keys(), averages.values(), color=['black'])
plt.title('Average Open, Close, High, and Low Prices')
plt.xlabel('Price Type')
plt.ylabel('Average Price')
plt.grid(True, axis='y')
plt.show()
```

```python
df1 = df['2013':'2014'].mean()['High']
df2 =df['2015':'2016'].mean()['High']
df3 =df['2017':'2018'].mean()['High']
df4 =df['2019':'2020'].mean()['High']
df5 =df['2021':'2022'].mean()['High']

periods = ['2013-2014', '2015-2016', '2017-2018', '2019-2020', '2021-2022']
averages = [df1, df2, df3, df4, df5]

# Create bar plot
plt.figure(figsize=(10, 6))
plt.bar(periods, averages, color='skyblue')
plt.title('Average High Prices for Specified Periods')
plt.xlabel('Period')
plt.ylabel('Average High Price')
plt.grid(True, axis='y')
plt.show()
```
